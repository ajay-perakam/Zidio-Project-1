```python
x=df.iloc[:,2:].values
x
```

```python
scale=MinMaxScaler()
features=scale.fit_transform(x)
features
```

```python
inertia=[]
for i in range(1,16):
  kmeans=KMeans(n_clusters=i)
  kmeans=kmeans.fit(features)
  inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.plot(range(1,16),inertia,'bo-')
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")


plt.subplot(1,2,2)
visualizer=KElbowVisualizer(KMeans(),k=(1,16))
visualizer.fit(features)
plt.suptitle("elbow graph and elbow visualizer")
visualizer.poof()
plt.show()
```

```python
silhouette_avg=[]
for i in range(2,16):
  kmeans=KMeans(n_clusters=i)
  cluster_labels=kmeans.fit_predict(features)
  silhouette_avg.append(silhouette_score(features,cluster_labels))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(2,16),silhouette_avg,'bX-')
plt.title("Silhouette analysis for optimal Score")
plt.xlabel("Values of K")
plt.ylabel("Silhouette score")
plt.show()
```

```python
model=KMeans(n_clusters=3)
model=model.fit(features)
```

```python
y_km=model.predict(features)
centers=model.cluster_centers_
```

```python
df['Cluster']=pd.DataFrame(y_km)
df.to_csv("Cluster_data",index=False)
```
